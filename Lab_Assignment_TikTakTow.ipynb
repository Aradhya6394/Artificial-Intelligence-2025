```python
"""
Tic Tac Toe using Minimax Search

OBJECTIVE:
1. Implement Tic Tac Toe game logic using Python.
2. Define state-space functions: S0, Player, Actions, Result, Terminal, Utility.
3. Implement Minimax algorithm to determine optimal moves.
4. Count and compare number of states explored vs theoretical.
"""

import itertools

# Step 1: Represent the initial state (S0)
S0 = [" " for _ in range(9)]  # Empty 3x3 board

#  STATE FUNCTIONS #

def Player(s):
    """Return which player moves next (X or O)."""
    X_count = s.count("X")
    O_count = s.count("O")
    return "X" if X_count == O_count else "O"

def Actions(s):
    """Return list of legal moves (empty cells)."""
    return [i for i in range(9) if s[i] == " "]

def Result(s, a):
    """Return the new state after applying action 'a'."""
    new_state = s.copy()
    new_state[a] = Player(s)
    return new_state

def check_winner(player, s):
    """Check if the given player has won in state s."""
    win_conditions = [
        [0,1,2],[3,4,5],[6,7,8],  # rows
        [0,3,6],[1,4,7],[2,5,8],  # cols
        [0,4,8],[2,4,6]           # diagonals
    ]
    for cond in win_conditions:
        if s[cond[0]] == s[cond[1]] == s[cond[2]] == player:
            return True
    return False

def Terminal(s):
    """Return True if game is over (win or draw)."""
    return check_winner("X", s) or check_winner("O", s) or " " not in s

def Utility(s):
    """Return +1 if X wins, -1 if O wins, 0 otherwise."""
    if check_winner("X", s): return 1
    elif check_winner("O", s): return -1
    else: return 0


# MINIMAX IMPLEMENTATION#

call_count = 0  # counts how many states AI evaluates

def MAX_VALUE(state):
    global call_count
    call_count += 1
    if Terminal(state):
        return Utility(state)
    v = float("-inf")
    for action in Actions(state):
        v = max(v, MIN_VALUE(Result(state, action)))
    return v

def MIN_VALUE(state):
    global call_count
    call_count += 1
    if Terminal(state):
        return Utility(state)
    v = float("inf")
    for action in Actions(state):
        v = min(v, MAX_VALUE(Result(state, action)))
    return v

def minimax_decision(state):
    """Return best move for current player and print number of states evaluated."""
    global call_count
    call_count = 0
    player = Player(state)

    if player == "X":  # Maximizer
        best_val = float("-inf")
        best_action = None
        for action in Actions(state):
            val = MIN_VALUE(Result(state, action))
            if val > best_val:
                best_val = val
                best_action = action
    else:  # Minimizer
        best_val = float("inf")
        best_action = None
        for action in Actions(state):
            val = MAX_VALUE(Result(state, action))
            if val < best_val:
                best_val = val
                best_action = action

    print(f"\nAI evaluated {call_count} states before choosing move {best_action + 1}")
    return best_action


# HELPER FUNCTIONS #

def print_board(s):
    """Display board in 3x3 format."""
    print(s[0]+"|"+s[1]+"|"+s[2])
    print("-+-+-")
    print(s[3]+"|"+s[4]+"|"+s[5])
    print("-+-+-")
    print(s[6]+"|"+s[7]+"|"+s[8])

def theoretical_states():
    """
    Return the theoretical maximum possible states in Tic Tac Toe:
    9! = 362,880 possible sequences of moves.
    But many states are symmetrical or impossible.
    Approx. ~26,830 valid unique states (known from research).
    """
    total = 9 * 8 * 7 * 6 * 5 * 4 * 3 * 2 * 1
    print(f"Theoretical maximum states (9!) = {total:,}")
    print("Approx. valid unique game states â‰ˆ 26,830\n")


# MAIN GAME LOOP #

def play_game():
    """Human (X) vs AI (O) using Minimax"""
    state = S0.copy()
    print("Welcome to Tic Tac Toe! (You are X, AI is O)")
    print("Positions are numbered 1â€“9 like this:\n")
    print("1|2|3")
    print("-+-+-")
    print("4|5|6")
    print("-+-+-")
    print("7|8|9\n")

    theoretical_states()

    while not Terminal(state):
        print_board(state)
        if Player(state) == "X":
            # Human move
            try:
                move = int(input("Enter your move (1-9): ")) - 1
                if move not in Actions(state):
                    print("Invalid move! Try again.")
                    continue
            except ValueError:
                print("Please enter a number between 1 and 9.")
                continue
        else:
            # AI move
            print("\nAI is thinking...")
            move = minimax_decision(state)
            print(f"AI chooses position {move + 1}\n")

        # Apply move
        state = Result(state, move)

    # Final result
    print_board(state)
    if Utility(state) == 1:
        print(" X wins!")
    elif Utility(state) == -1:
        print("O wins!")
    else:
        print(" It's a draw!")

# RUN #
if __name__ == "__main__":
    play_game()
```
