```python
"""
TIC-TAC-TOE USING MINIMAX ALGORITHM
TOPIC:
Implementation of Tic-Tac-Toe AI using the **Minimax Algorithm**
â€” a classic example of adversarial search in Artificial Intelligence.

ABOUT THE ALGORITHM:
Minimax is a **backtracking algorithm** used in decision-making and
game theory. It assumes that both players play optimally.

- 'X' â†’ Maximizing player (tries to maximize score)
- 'O' â†’ Minimizing player (tries to minimize score)

  Minimax explores all possible future moves (game tree)
  Assigns scores:
    +1 â†’ X wins
    -1 â†’ O wins
     0 â†’ Draw
  X chooses the move that gives the maximum guaranteed score
  O chooses the move that gives the minimum guaranteed score

  FEATURES:
- Human (X) vs Computer (O)
- AI uses full Minimax search (no depth limit)
- Explains how many states AI evaluated before making a decision
- Well-commented for easy understanding


"""

# ---------------- STEP 1: INITIAL STATE ----------------
S0 = [" " for _ in range(9)]   # empty 3x3 board (represented as list of 9 cells)


# ---------------- STEP 2: BASIC GAME FUNCTIONS ----------------
def Player(s):
    """
    Return which player's turn it is.
    - X always starts first.
    - If equal number of X and O â†’ X's turn.
    - Otherwise â†’ O's turn.
    """
    X_count = s.count("X")
    O_count = s.count("O")
    return "X" if X_count == O_count else "O"


def Actions(s):
    """Return list of available actions (empty cell indices)."""
    return [i for i in range(9) if s[i] == " "]


def Result(s, a):
    """
    Return new board state after applying action 'a'.
    Does not modify the original state.
    """
    new_state = s.copy()
    new_state[a] = Player(s)
    return new_state


def check_winner(player, s):
    """Check whether the given player (X or O) has a winning combination."""
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # columns
        [0, 4, 8], [2, 4, 6]              # diagonals
    ]
    for cond in win_conditions:
        if s[cond[0]] == s[cond[1]] == s[cond[2]] == player:
            return True
    return False


def Terminal(s):
    """
    Check whether the game has ended.
    True if:
    - X or O has won, OR
    - The board is full (Draw).
    """
    return check_winner("X", s) or check_winner("O", s) or " " not in s


def Utility(s):
    """
    Utility function:
    +1 â†’ X wins
    -1 â†’ O wins
     0 â†’ Draw or ongoing
    """
    if check_winner("X", s):
        return 1
    elif check_winner("O", s):
        return -1
    else:
        return 0


# ---------------- STEP 3: MINIMAX ALGORITHM ----------------
call_count = 0  # to count how many recursive calls are made


def MAX_VALUE(state):
    """Compute maximum value for X (maximizing player)."""
    global call_count
    call_count += 1

    # Base case: stop if game over
    if Terminal(state):
        return Utility(state)

    v = float("-inf")  # initialize best score to negative infinity
    for action in Actions(state):
        v = max(v, MIN_VALUE(Result(state, action)))
    return v


def MIN_VALUE(state):
    """Compute minimum value for O (minimizing player)."""
    global call_count
    call_count += 1

    if Terminal(state):
        return Utility(state)

    v = float("inf")  # initialize best score to positive infinity
    for action in Actions(state):
        v = min(v, MAX_VALUE(Result(state, action)))
    return v


# ---------------- STEP 4: DECISION FUNCTION ----------------
def minimax_decision(state):
    """
    Determine the best move for the current player using Minimax.
    Also reports how many states were evaluated.
    """
    global call_count
    call_count = 0  # reset counter each time
    player = Player(state)

    # If it's X's turn â†’ maximize
    if player == "X":
        best_val = float("-inf")
        best_action = None
        for action in Actions(state):
            val = MIN_VALUE(Result(state, action))
            if val > best_val:
                best_val = val
                best_action = action

    # If it's O's turn â†’ minimize
    else:
        best_val = float("inf")
        best_action = None
        for action in Actions(state):
            val = MAX_VALUE(Result(state, action))
            if val < best_val:
                best_val = val
                best_action = action

    print(f"AI evaluated {call_count} states before choosing move {best_action + 1}")
    return best_action


# ---------------- STEP 5: DISPLAY FUNCTION ----------------
def print_board(s):
    """Print the current Tic-Tac-Toe board."""
    print(s[0] + "|" + s[1] + "|" + s[2])
    print("-+-+-")
    print(s[3] + "|" + s[4] + "|" + s[5])
    print("-+-+-")
    print(s[6] + "|" + s[7] + "|" + s[8])


# ---------------- STEP 6: MAIN GAME LOOP ----------------
state = S0.copy()
print("Welcome to Tic-Tac-Toe! (You are X, AI is O)")
print("Positions are numbered 1â€“9 like this:")
print("1|2|3")
print("-+-+-")
print("4|5|6")
print("-+-+-")
print("7|8|9\n")

# Play until the game reaches a terminal state
while not Terminal(state):
    print_board(state)

    # HUMAN TURN (X)
    if Player(state) == "X":
        try:
            move = int(input("Enter your move (1-9): ")) - 1
            if move not in Actions(state):
                print("Invalid move! Try again.")
                continue
        except ValueError:
            print("Please enter a number between 1 and 9.")
            continue

    # AI TURN (O)
    else:
        print("AI is thinking...")
        move = minimax_decision(state)
        print(f"AI chooses position {move + 1}")

    # Apply chosen move to the game state
    state = Result(state, move)

# ---------------- GAME OVER ----------------
print_board(state)
if Utility(state) == 1:
    print("X wins!")
elif Utility(state) == -1:
    print("O wins!")
else:
    print("It's a draw!")
```
